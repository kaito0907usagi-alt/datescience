```python
def binomial_probability(n, x, p):
    # nCx を計算 (n! / (x! * (n-x)!))
    combination = math.factorial(n) / (math.factorial(x) * math.factorial(n - x))
    # 二項分布の確率を計算 P(X=x)
    probability = combination * (p ** x) * ((1 - p) ** (n - x))
    return probability

n = 20
p = 0.6
probabilities = [binomial_probability(n, x, p) for x in range(n+1)]
x_values = np.arange(0, n+1)

plt.figure(figsize=(6, 5))
plt.stem(x_values, probabilities, 'black', markerfmt='', basefmt=' ')

for i, prob in enumerate(probabilities):
    color = 'black'
    plt.text(i, prob + 0.01, f"{prob:.4f}", ha='center', fontsize=15, rotation=90, color=color)

plt.xticks(np.arange(0, n + 1, 4))

plt.ylim(ymin=0, ymax=0.25)
plt.xlabel("$x$", fontsize=20, labelpad=8, math_fontfamily='cm')
plt.ylabel("発生確率", fontsize=14)
plt.tick_params(labelsize=12)

plt.tight_layout()
plt.show()
```

<span style="font-size: 20px; color: #F08000;">例題　二項検定</span>

定評のあるダイエット機器Aがあり，使用者の $60$% に効果があることがわかっている

ある学生らは新しいダイエット機器Bを開発したが，その有効率は現時点では不明である

その有効性を確かめる実験として，$20$人の被検者を集め，何人に効果があるか確かめた

機器Bを使用した $20$人うち，$18$人（$90$%）に効果があったとする

この結果から機器Bは機器Aより優れていると判断できるだろうか

二項検定で確かめよ．ただし有意水準は $5$% とすること

```python
n = 20                   # 試行回数（被験者数）
observed_successes = 18  # 機器Bで効果があった人数
p_null = 0.60            # 帰無仮説の成功確率（機器Aの有効率）

# 二項検定（片側検定：効果が優れているかを確認）
p_value = 1 - stats.binom.cdf(observed_successes - 1, n, p_null) 

alpha = 0.05
is_significant = p_value < alpha

# 結果の表示
print(f"観測された成功回数: {observed_successes}")
print(f"帰無仮説の成功確率: {p_null:.2f}")
print(f"p値: {p_value:.4f}")

if is_significant:
    print("結果: 統計的に有意です（帰無仮説を棄却します）．")
    print("結論: 機器Bは機器Aより優れていると判断できます．")
else:
    print("結果: 統計的に有意ではありません（帰無仮説を棄却できません）．")
    print("結論: 機器Bが機器Aより優れているとは判断できません．")
```
